```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text  import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('synthetic_text_data.csv')
```

```python
x= df['text']
y=df['label']
```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y,random_state=32,test_size=0.2)
```

```python
vectonizer = CountVectorizer()
X_train_vect = vectonizer.fit_transform(x_train)
X_test_vect = vectonizer.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB().fit(X_train_vect,y_train)
```

```python
y_preds = model.predict(X_test_vect)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
acc_score = accuracy_score(y_test,y_preds)
print('Accuracy of model is: {}'.format(acc_score*100))
```

```python
cf = confusion_matrix(y_test,y_preds)
```
